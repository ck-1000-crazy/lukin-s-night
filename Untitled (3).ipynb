```python
import random

item_descriptions = {
    "фонарь": "Старый керосиновый фонарь. Даёт тусклый, но настоящий свет.",
    "амулет": "Подвеска с глазом. Легенда: отгоняет взгляд Лукина.",
    "фигурка лукин": "Глиняная фигурка гоблина. Её маленькие глаза словно живые.",
    "клочок ткани": "Пыльный кусочек ткани с неясными пятнами",
    "кукла": "Тряпичная кукла. Она просто смотрит. И ты решаешь не проверять — чем."
}

player = {
    "name": None,
    "intensity": 0,
    "sanity": 10
}

hyperrealitys = {
    "start": {
        "text": "Нижняя Поповка. Ночь. Дом с выцарапанным словом 'ЛУКИН' стоит на холме.",
        "options": [
            {"text": "Подойти к дому и войти", "next": "porch"},
            {"text": "Осмотреть окрестности", "next": "outside"},
            {"text": "Уйти прочь", "next": "ending_escape", "ending": "escape"}
        ]
    },
    "outside": {
        "text": "В зарослях ты находишь потертый амулет и фонари.",
        "options": [
            {"text": "Взять фонарь", "next": "outside", "add_item": "фонарь"},
            {"text": "Взять амулет", "next": "outside", "add_item": "амулет", "add_flag": "take_amulet"},
            {"text": "Вернуться к крыльцу", "next": "porch"}
        ],
        "effect": {"intensity": +1}
    },
    "porch": {
        "text": "Крыльцо. Дверь приоткрыта. Тишина кажется слишком густой.",
        "options": [
            {"text": "Войти в дом", "next": "hall"},
            {"text": "Попробовать тихо открыть окно", "next": "window_peek"},
            {"text": "Шепнуть в дом", "next": "caught_steal", "add_flag": "tried_steal"}
        ],
        "effect": {"intensity": +1}
    },
    "window_peek": {
        "text": "Через запылённое окно виднеется фигурка на столе.",
        "options": [
            {"text": "Постучать", "next": "window_knock"},
            {"text": "Вернуться и войти по двери", "next": "hall"}
        ],
        "effect": {"intensity": +1}
    },
    "window_knock": {
        "text": "Кто-то шепчет твоё имя. Дверь слегка захлопнулась.",
        "options": [
            {"text": "Войти внутрь", "next": "hall"},
            {"text": "Уйти (тебя трясёт)", "next": "ending_faint", "ending": "faint"}
        ],
        "effect": {"sanity": -1}
    },
    "caught_steal": {
        "text": "Тебя заперли внутри. Прихожая впереди.",
        "options": [
            {"text": "Осмотреть прихожую", "next": "hall"},
            {"text": "Крикнуть на помощь", "next": "ending_taken", "ending": "taken_by_dark"}
        ],
        "effect": {"intensity": +1}
    },
    "hall": {
        "text": "Прихожая. Лестница наверх и дверь в подвал.",
        "options": [
            {"text": "Подняться наверх", "next": "attic"},
            {"text": "Спуститься в подвал", "next": "basement_entrance"},
            {"text": "Осмотреть стол", "next": "table"}
        ],
        "effect": {"intensity": +1}
    },
    "table": {
        "text": "На столе тряпичная кукла и глиняная фигурка Лукина.",
        "options": [
            {"text": "Взять фигурку", "next": "take_figurine", "add_item": "фигурка лукин"},
            {"text": "Разбить фигурку", "next": "break_figurine", "add_flag": "broke_figurine"},
            {"text": "Вернуться", "next": "hall"}
        ],
        "effect": {"intensity": +1}
    },
    "tаble": {
        "text": "На столе снова стоит глиняная фигурка Лукина. Тряпичной куклы нет. Хмм... похоже, здесь чего-то не хватает.",
        "options": [
            {"text": "Взять фигурку", "next": "hall", "add_item": "фигурка лукин"},
            {"text": "Разбить фигурку", "next": "break_figurine", "add_flag": "broke_figurine"},
            {"text": "Вернуться", "next": "hall"}
        ],
        "effect": {"intensity": +1}
    },
    "take_figurine": {
        "text": "Прихожая. Лестница наверх и дверь в подвал.",
        "options": [
            {"text": "Подняться наверх", "next": "attic"},
            {"text": "Спуститься в подвал", "next": "basement_entrance"},
        ],
        "effect": {"intensity": +1}
    },
      "break_figurine": {
        "text": "Фигурка рассыпалась; в стенах что-то зашевелилось.",
        "options": [
            {"text": "Подняться наверх", "next": "attic"},
            {"text": "Спуститься в подвал", "next": "basement_entrance"},
        ],
        "effect": {"intensity": +1}
    },
    "basement_entrance": {
        "text": "Вход в подвал. Холод тянется вверх.",
        "options": [
            {"text": "Сойти наощупь вниз (риск)", "next": "pitfall"},
            {"text": "Зажечь фонарь и спуститься", "next": "basement", "requires_item": "фонарь"},
            {"text": "Вернуться в прихожую", "next": "hall"}
        ],
        "effect": {"intensity": +1}
    },
    "pitfall": {
        "text": "Ты спотыкаешься и падаешь в скрытую ямку. Внизу маленький глаз мерцает.",
        "options": [
            {"text": "Пытаться выбраться", "next": "struggle"},
            {"text": "Позвать на помощь", "next": "ending_taken", "ending": "taken_by_dark"}
        ],
        "effect": {"intensity": +1}
    },
    "struggle": {
        "text": "Ты выбираешься, но рассудок ослаб. Тебя выкидывает у порога дома.",
        "options": [
            {"text": "Подняться наверх", "next": "attic"},
            {"text": "Войти в прихожую снова", "next": "hall"},
            {"text": "Убежать из Нижней Поповки", "next": "ending_escape", "ending": "escape"}
        ],
        "effect": {"sanity": -3}
    },
    "basement": {
        "text": "В подвале фонарь освещает тёмный кристалл на плите.",
        "options": [
            {"text": "Разбить кристалл", "next": "destroy_crystal"},
            {"text": "Положить амулет на кристалл", "next": "use_amulet", "requires_item": "амулет"},
            {"text": "Схватить кристалл и бежать", "next": "take_crystal", "add_flag": "took_crystal"}
        ],
        "effect": {"intensity": +1}
    },
    "use_amulet": {
        "text": "Амулет сияет, кристалл ослаб. Голос Лукина вздыхает.",
        "options": [
            {"text": "Уничтожить кристалл сейчас", "next": "destroy_crystal"},
            {"text": "Остаться и наблюдать", "next": "ending_peace", "ending": "peace"}
        ],
        "effect": {"intensity": +1}
    },
    "destroy_crystal": {
        "text": "Кристалл разбит. Дом предлагает: остаться хранителем или уйти с фонарём.",
        "options": [
            {"text": "Остаться и оставить фонарь", "next": "ending_keeper", "ending": "keeper", "add_item": "фонарь"},
            {"text": "Унести фонарь и бежать", "next": "ending_peace", "ending": "peace", "add_flag": "saved_village"}
        ],
        "effect": {"intensity": +1}
    },
    "take_crystal": {
        "text": "Кристалл прожёг ладонь; голос внутри: 'Найди глаза'.",
        "options": [
            {"text": "Подчиниться голосу", "next": "offer_eyes"},
            {"text": "Разбить кристалл", "next": "destroy_crystal"}
        ],
        "effect": {"sanity": -5}
    },
    "offer_eyes": {
        "text": "Кристалл требует жертв и указывает на кровать на чердаке.",
        "options": [
            {"text": "Подчиниться и принести жертву", "next": "ending_corrupt", "ending": "corrupt"},
            {"text": "Отклонить и разбить кристалл", "next": "destroy_crystal"}
        ],
        "effect": {"intensity": +1}
    },
    "attic": {
        "text": "Чердак тёмный. В углу — сундук и кровать. Здесь можно найти подсказки.",
        "options": [
            {"text": "Открыть сундук", "next": "chest"},
            {"text": "Подойти к кровати и осмотреть", "next": "attic_bed"},
            {"text": "Вернуться в прихожую", "next": "hall"}
        ],
        "effect": {"intensity": +1}
    },
    "chest": {
        "text": "Ты залезаешь в сундук. Внутри — бесконечные клочки ткани, будто сундук шьётся изнутри.",
        "options": [
            {"text": "Взять клочок ткани", "next": "chest", "add_item": "клочок ткани"},
            {"text": "Выбраться из сундука", "next": "attic"}
        ]
    },
    "attic_bed": {
        "text": "Кровать покрыта старым одеялом. Под ним что-то шевелится.",
        "options": [
            {"text": "Сорвать одеяло (рискованно)", "next": "lukin_under_blanket_check"},
            {"text": "Поднести амулет и посмотреть (если есть)", "next": "amulettest", "requires_item": "амулет"},
            {"text": "Вернуться", "next": "attic"}
        ],
        "effect": {"intensity": +1}
    },
    "amulettest": {
        "text": "Амулет вибрирует; шёпоты притихают. Можно сорвать одеяло с меньшим риском.",
        "options": [
            {"text": "Сорвать одеяло", "next": "lukin_under_blanket_check"},
            {"text": "Вернуться", "next": "attic"}
        ],
        "effect": {"intensity": +1}
    },
    "lukin_under_blanket_check": {
        "text": "Ты срываешь одеяло. Там — маленький гоблин Лукин с огромными глазами.",
        "options": [
            {"text": "Схватить Лукина за нос и попытаться понять шёпоты", "next": "ending_lukin_blanket", "ending": "lukin_blanket"},
            {"text": "Положить фигурку рядом и закрыть одеяло", "next": "ending_buried", "ending": "buried", "requires_item": "фигурка лукин"}
        ],
        "effect": {"intensity": +1}
    },
    "doll_event": {
        "text": "Ты чувствуешь, как в кармане что-то мягко шевелится. Кукла уже там. Теперь у неё есть глаза. Они моргают.",
        "options": [
            {"text": "Посмотреть в глаза", "next": "ending_doll_eyes", "ending": "doll_eyes"},
            {"text": "Сжечь куклу (если есть фонарь)", "next": "ending_burn_doll", "requires_item": "фонарь"},
            {"text": "Убежать", "next": "escape_doll"}
        ],
        "effect": {"intensity": +1}
    },
    "escape_doll": {
        "text": "Ты бросаешься к выходу, но дверь заперта.",
        "options": [
            {"text": "Подняться наверх", "next": "ending_taken_by_cloth"},
            {"text": "Спуститься в подвал", "next": "ending_taken_by_cloth"},
            {"text": "Попытаться вспомнить, как ты сюда попал", "next": "memory_gap"}
        ],
        "effect": {"intensity": +1}
    },

    "ending_taken": {"text": "Дом поглотил тебя. Твои глаза теперь в рамке.", "options": []},
    "ending_taken_by_cloth": {"text": "Ты пытаешься двинуть ноги. Но между ними уже нет пространства. Только ткань.\n"
                            "Она просачивается сквозь пальцы глаз - похоже ей не нравится, что её трогают.", "options": []},
    "ending_keeper": {"text": "Ты совершил побратание с Лукиным. Теперь ты хранитель. Ночь слушает твои шаги.", "options": []},
    "ending_peace": {"text": "Ты спас Нижнюю Поповку.", "options": []},
    "ending_corrupt": {"text": "Ты стал требовать жертв.", "options": []},
    "ending_lukin_blanket": {"text": "Лукин поджидает тебя под одеялком. Он смотрит прямо в глаза.", "options": []},
    "ending_buried": {"text": "Ты похоронил Лукина под одеялком. Его шёпот тих.", "options": []},
    "ending_escape": {"text": "Ты ушёл и спас свою жизнь.", "options": []},
    "ending_faint": {"text": "Ты упал в обморок и очнулся в другом месте.", "options": []},
    "ending_doll_eyes": {"text": "Ты смотришь в глаза куклы. Теперь они у тебя одни на двоих.", "options": []},
    "ending_burn_doll": {"text": "Кукла горит без огня. Дом трещит. Ты выбегаешь наружу. Но запах ткани остаётся с тобой.", "options": []},
    "ending_lukin_secret": {"text": "Добро пожаловать домой. Грибные хинкали уже поджидают тебя.", "options": []},
}

inventory = []
flags = set()
history = []
figurine_breaks = 0
amulets = 0

def display_hyperreality(key):
    hyperreality = hyperrealitys.get(key)
    if hyperreality is None:
        print("Ошибка: сцена не найдена.")
        return None
    print("\n=== {} ===".format(key))
    print(hyperreality["text"])
    opts = hyperreality.get("options", [])
    if not opts:
        return hyperreality
    for i, opt in enumerate(opts, start=1):
        print(f"{i}. {opt['text']}")
    return hyperreality

def maybe_whisper():
    i = player["intensity"]
    if i >= 20 and random.random() < i*0.01:
        fake = random.choice([
            "Выберите номер варианта: 1",
            "Инвентарь обновлён: ты",
            "Сохранение...",
            "Под ногами тёмные каракули с глазами"
            "Лукин проснулся"])
        print(fake)

def apply_option(current_key, index):
    hyperreality = hyperrealitys.get(current_key)
    if hyperreality is None:
        return current_key
    opts = hyperreality.get("options", [])
    if not opts:
        return current_key
    if index < 1 or index > len(opts):
        print("Неверный номер варианта.")
        return current_key
    opt = opts[index - 1]
    history.append(f"{current_key}: {opt['text']}")
    if opt.get("requires_item"):
        req = opt["requires_item"]
        if req not in inventory:
            print(f"Для этого варианта требуется: {req}.")
            return current_key
        else:
            inventory.remove(req)
            print(f"Предмет {req} использован.")

    if opt.get("add_item"):
        item = opt["add_item"]
        if item == "клочок ткани":
            inventory.append(item)
            print("Добавлен предмет: клочок ткани")
        elif item not in inventory:
            inventory.append(item)
            print(f"Добавлен предмет: {item}")
        else:
            print("Достаточно")

    if opt.get("add_flag"):
        flags.add(opt["add_flag"]) 
        
    global figurine_breaks
    if opt.get("add_flag") == "broke_figurine":
        figurine_breaks += 1
        hyperrealitys["hall"]["options"] = [
            {"text": "Подняться наверх", "next": "attic"},
            {"text": "Спуститься в подвал", "next": "basement_entrance"},
            {"text": "Осмотреть стол", "next": "tаble"}]

    global amulets
    if opt.get("add_flag") == "take_amulet":
        amulets += 1
        hyperrealitys["outside"]["options"] = [
            {"text": "Взять фонарь", "next": "outside", "add_item": "фонарь"},
            {"text": "Вернуться к крыльцу", "next": "porch"}]
        hyperrealitys["outside"]["text"] = "В зарослях остались только фонари."
    
    maybe_spawn_doll()
        
    eff = hyperreality.get("effect")
    if eff and isinstance(eff, dict):
        if "sanity" in eff:
            player["sanity"] += eff["sanity"]
            if player["sanity"] > 10:
                player["sanity"] = 10
            print(f"Рассудок: {player['sanity']}")
        player["intensity"] += 1
        if player["intensity"] < 0:
            player["intensity"] = 0
        print(f"Интенсивность: {player['intensity']}")
            
    nxt = opt.get("next")
    if not nxt:
        return current_key
    if nxt not in hyperrealitys:
        print("Путь недоступен, остаёшься здесь.")
        return current_key
    if figurine_breaks >= 3:
        return "doll_event"
    if nxt == "porch" and "is_lukin" in flags:
        return "ending_lukin_secret"
    return nxt

def save_results(filename="results.txt"):
    try:
        with open(filename, "w", encoding="utf-8") as f:
            f.write("Прохождение: Нижняя Поповка: Ночь Лукина\n")
            f.write(f"Игрок: {player.get('name','Unknown')}\n\n")
            f.write("Ход игры:\n")
            for i, h in enumerate(history, start=1):
                f.write(f"{i}. {h}\n")
            f.write("\nИнвентарь:\n")
            if inventory:
                for it in inventory:
                    f.write(f"- {it}: {item_descriptions.get(it,'Нет описания')}\n")
            else:
                f.write("Пусто\n")
            f.write("\nФлаги:\n")
            if flags:
                for fl in flags:
                    f.write(f"- {fl}\n")
            else:
                f.write("Нет\n")
            f.write(f"\nСостояние игрока: intensity={player['intensity']}, sanity={player['sanity']}\n")
        print(f"Результаты сохранены в {filename}")
    except Exception as e:
        print("Ошибка при записи файла:", e)
        
def maybe_spawn_doll():
    if "кукла" not in inventory:
        chance = 0.34 * figurine_breaks + 0.03 * inventory.count("клочок ткани")
        if random.random() < chance:
            inventory.append("кукла")
            print("Добавлен предмет: кукла")

def main():
    print("Добро пожаловать: Нижняя Поповка: Ночь Лукина")
    name = input("Введите имя: ").strip()
    player["name"] = name if name else "Путник"
    current = "start"
    ending = None
    lukin_names = {"ЛУКИН", "LUKIN", "LU", "ЛУ"}
    if player["name"].upper() in lukin_names:
        flags.add("is_lukin")
    while True:
        hyperreality = display_hyperreality(current)
        if hyperreality is None:
            print("Ошибка сцены. Завершение.")
            break

        if not hyperreality.get("options"):
            ending = current
            history.append(f"Концовка: {hyperreality['text']}")
            print("\n--- Концовка ---")
            print(hyperreality['text'])
            break
        while True: 
            choice_raw = input("Выберите номер варианта: ").strip()
            if not choice_raw.isdigit():
                print("Введи номер варианта (число).")
                continue
            choice = int(choice_raw)
            break

        next_key = apply_option(current, choice)
        maybe_whisper()
        
        if next_key is None:
            print("Неверный переход. Попробуйте снова.")
            continue

        current = next_key
        if player["sanity"] <= 0:
            ending = "ending_taken"
            history.append("Рассудок упал до нуля")
            print("Рассудок разрушен. Тьма победила.")
            break
        elif player["intensity"] >= 47:
            ending = "ending_taken"
            history.append("Интенсивность сквозит через мозаику")
            print("На него смотрит пустота, просачиваясь чистой интенсивностью через маленькие надломы.")
            break

    if not ending:
        ending = current

    history.append(f"Финальная сцена: {ending}")
    save_results("results.txt")
    print("Игра окончена. Результаты сохранены в results.txt")

if __name__ == "__main__":
    main()
```
